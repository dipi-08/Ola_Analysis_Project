```python
# Donut Chart : To find the ratio of booking status in each slices.

import matplotlib.pyplot as plt

file = "D:/Annudip Notes/Data/DATA ANALYSIS PROJECT/Bookings.csv"
data = pd.read_csv(file)

# Grouping two columns to make the booking status
comparison = data.groupby('Booking_Status')['Booking_Value'].sum()

# Extract data from the group to make donut chart
booking_status = comparison.index
booking_value = comparison.values

# Defining the color palette
colors = plt.cm.Paired(range(len(booking_status)))

# Creating the donut chart 
fig, ax = plt.subplots(figsize=(6,6)) 
wedges, texts, autotexts = ax.pie(booking_value, labels=booking_status, autopct='%1.1f%%', startangle=140, wedgeprops={'width':0.6}, colors=colors)

# Create a legend
handles =[plt.Rectangle((0,0), 1,1, color=color) for color in colors]
plt.legend(handles, booking_status, title='Booking Status', bbox_to_anchor=(1.05, 1), loc='upper left')

# Setting the title and showing the chart
plt.title('Booking Status Breakdown')
plt.show()
```

```python
# Boxplot or Box-and-Whisker Plot : The diagram is used for finding the minimum and maximum range of the vehicles as well as finding the value of first and third quratiles alongwith median. 
# Features of a Boxplot:
# Box: The box represents the interquartile range (IQR), i.e., the range between Q1 and Q3.
# Whiskers: Lines extending from the box that represent data within 1.5 times the IQR from Q1 and Q3. Any data points outside this range are considered outliers and are marked separately.
# Outliers: Points outside the whiskers are considered outliers, often marked as individual points.
# Median Line: A line inside the box that indicates the median value of the data.

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = "D:/Annudip Notes/Data/DATA ANALYSIS PROJECT/Bookings.csv"
df = pd.read_csv(file)

# Check for the correct columns
plt.figure(figsize=(10, 5))

# Assuming 'Booking_Value' is a numerical column for the y-axis
ax = sns.boxplot(x='Vehicle_Type', y='Booking_Value', data=df, hue='Booking_Status', palette='Set2')

# Correct axis labels
plt.xlabel('Vehicle Type')  # Updated to reflect the data
plt.ylabel('Booking Value')   # Assuming you're plotting booking value (change if you have another numerical column)
plt.title('Boxplot Chart Comparison of Vehicle Type and Booking Status')

# Adjust the legend to be placed outside the plot
plt.legend(title='Booking Status', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

# Ensure everything fits within the figure area
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Waterfall Chart : Basically it is useful for visualizing incremental changes to a starting value, showing how different factors contribute to an overall change.
# A waterfall chart typically works by visualizing "positive" or "negative" changes from an initial value.
# Therefore, by using this we can know the changes happened during the time of booking and according to it how much is the success rate.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = "D:/Annudip Notes/Data/DATA ANALYSIS PROJECT/Bookings.csv"
df = pd.read_csv(file)

# Group the data by 'Booking_Status' and count the number of 'Booking_ID' in each status
df_grouped = df.groupby('Booking_Status')['Booking_ID'].count().reset_index()

# Sort the 'Booking_Status' if you want to control the order in the waterfall chart
df_grouped = df_grouped.sort_values(by='Booking_ID', ascending=False)

# Calculate the cumulative sum to determine how each status contributes to the overall total
df_grouped['Cumulative'] = df_grouped['Booking_ID'].cumsum()

# To create the waterfall chart, we'll add a starting point and calculate each step
initial_value = df_grouped['Booking_ID'].iloc[0]  # The first value will be the starting point
df_grouped['Waterfall'] = df_grouped['Booking_ID'] - initial_value

# Create a waterfall chart
plt.figure(figsize=(10, 6))

# Generate the bars for the waterfall chart
plt.bar(df_grouped['Booking_Status'], df_grouped['Waterfall'], color='skyblue')

# Add starting point
plt.bar(df_grouped['Booking_Status'].iloc[0], initial_value, color='lightgreen')

# Add labels for each bar showing cumulative values
for i in range(len(df_grouped)):
    plt.text(i, df_grouped['Waterfall'].iloc[i] + initial_value, str(df_grouped['Booking_ID'].iloc[i]),
             ha='center', va='bottom')

# Adding title and labels
plt.title('Waterfall Chart for Booking ID by Booking Status')
plt.xlabel('Booking Status')
plt.ylabel('Change in Booking ID')

plt.tight_layout()
plt.show()
```
